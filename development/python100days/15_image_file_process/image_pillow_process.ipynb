```python
rect = (200,200,600,600)
img_crop = img.crop(rect)
```

```python
img_crop.show()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure()
plt.imshow(img_crop)
plt.axis('off')
plt.show()
```

```python
img.thumbnail((128,128))
img.show()
```

```python
img = Image.open('./garden-at-sainte-adresse.jpg')
for x in range(200,601):
    for y in range(200,601):
        img.putpixel((x,y),(255,255,255))
plt.figure()
plt.imshow(img)
plt.show()
# img.show()
```

```python
from PIL import ImageFilter
```

```python
img = Image.open('./garden-at-sainte-adresse.jpg')
img_filter = img.filter(ImageFilter.CONTOUR)
plt.figure(figsize=(8,6))
plt.imshow(img_filter)
plt.show()
```
